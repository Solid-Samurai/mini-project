```python
import torch
import torchvision.models as models
import torchvision.transforms as transform
from torch import nn, optim
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
# Load the original image and its corresponding mask
image_path = "Untitled Folder/image/id_1_value_13_116.jpg"
mask_path = "Untitled Folder/mask/id_1_value_13_116.jpg"  # Mask that encloses the meter reading

image = cv2.imread(image_path)
mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)  # Load mask in grayscale

# Apply the mask to extract the meter reading
masked_region = cv2.bitwise_and(image, image, mask=mask)

# Show the extracted region
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(masked_region, cv2.COLOR_BGR2RGB))
plt.title("Masked Water Meter Region")
plt.axis("off")
plt.show()
```

```python
# Convert to grayscale
gray = cv2.cvtColor(masked_region, cv2.COLOR_BGR2GRAY)

# Apply Otsu Thresholding
#thresh = cv2.bitwise_not(thresh)

#_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Alternative: Manually set a threshold (adjust 150 if needed)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

# Show the thresholded image
plt.figure(figsize=(6, 6))
plt.imshow(thresh, cmap="gray")
plt.title("Thresholded Masked Region")
plt.axis("off")
plt.show()

```

```python
# Find contours of digits
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

digit_regions = []

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    
    # Filtering: Ignore very small/large contours
    if 20 < w < 100 and 40 < h < 150:  
        digit_regions.append((x, y, w, h))

# Sort digit regions from left to right
digit_regions = sorted(digit_regions, key=lambda x: x[0])

# Draw bounding boxes around detected digits
digit_image = masked_region.copy()
for (x, y, w, h) in digit_regions:
    cv2.rectangle(digit_image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Show detected digits
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(digit_image, cv2.COLOR_BGR2RGB))
plt.title("Detected Digits in Masked Region")
plt.axis("off")
plt.show()
```

```python


# Load Pretrained ResNet-18
model = models.resnet18(pretrained=True)

# Modify Last Layer (10 classes for digits 0-9)
num_classes = 10  # Digits 0-9
model.fc = nn.Linear(in_features=model.fc.in_features, out_features=num_classes)

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
model.eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to 224x224
    transforms.Grayscale(num_output_channels=3),  # Convert to 3-channel grayscale
    transforms.ToTensor(),  # Convert image to tensor
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize
])

recognized_digits = []
for i, (x, y, w, h) in enumerate(digit_regions):
    digit_crop = thresh[y:y+h, x:x+w]  # Crop the digit region
    digit_crop = cv2.resize(digit_crop, (224, 224))  # Resize for ResNet-18
    digit_crop = transform(digit_crop).unsqueeze(0).to(device)  # Apply transformations

    # Predict using ResNet-18
    with torch.no_grad():
        output = model(digit_crop)
        predicted_class = torch.argmax(output, dim=1).item()

    recognized_digits.append(str(predicted_class))

# Print final water meter reading
final_reading = "".join(recognized_digits)
print(f"Recognized Water Meter Reading: {final_reading}")
```

```python
from torch.utils.data import DataLoader, Dataset
from torchvision.datasets import ImageFolder

# Load dataset (replace "data/" with actual dataset folder)
dataset = ImageFolder(root="Untitled Folder/image/", transform=transform)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

# Define Loss & Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train Model
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}")

print("Training Complete!")
```

```python
# Load validation dataset
val_dataset = ImageFolder(root="Untitled Folder/Testing/", transform=transform)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Class labels mapping
class_names = val_dataset.classes  # ['0', '1', '2', ..., '9']
print(f"Classes: {class_names}")
```

```python
correct = 0
total = 0

with torch.no_grad():  # Disable gradients for validation
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        _, predicted = torch.max(outputs, 1)  # Get predicted class

        correct += (predicted == labels).sum().item()
        total += labels.size(0)

accuracy = (correct / total) * 100
print(f"Validation Accuracy: {accuracy:.2f}%")
```

```python
import random

# Get a batch of images
dataiter = iter(val_loader)
images, labels = next(dataiter)
images, labels = images.to(device), labels.to(device)

# Get predictions
outputs = model(images)
_, predicted = torch.max(outputs, 1)

# Convert tensors to NumPy for visualization
images_np = images.cpu().numpy().transpose((0, 2, 3, 1))  # Convert (C, H, W) → (H, W, C)

# Plot first 5 test images with predictions
fig, axes = plt.subplots(1, 5, figsize=(12, 6))
for i in range(5):
    ax = axes[i]
    ax.imshow(images_np[i], cmap="gray")
    ax.set_title(f"Pred: {predicted[i].item()}, True: {labels[i].item()}")
    ax.axis("off")

plt.show()
```
