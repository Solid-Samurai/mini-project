```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
```

```python
water_usage=pd.read_csv('Water Usage/SOC Outlet-1/Table.csv',usecols=["Date","Time","Device Name","water_flow_in_cubic_meter"],parse_dates=["Date"],index_col="Date")
data=water_usage[::-1]
data.reset_index(inplace=True)
data.head(20)
```

```python
# Normalize the water consumption data
scaler = MinMaxScaler(feature_range=(0, 1))
data["water_flow_in_cubic_meter"] = scaler.fit_transform(data[["water_flow_in_cubic_meter"]])

# Convert data to NumPy array
data_values = data["water_flow_in_cubic_meter"].values
```

```python
def create_sequences(data, time_steps=10):
    """ Convert data into sequences of past `time_steps` readings as input (X) and the next reading as target (y). """
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i : i + time_steps])  # Past `time_steps` values as input
        y.append(data[i + time_steps])      # Predict next value

    X, y = np.array(X), np.array(y)

    # ✅ Ensure 3D shape: (samples, time_steps, features=1)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    return X, y

# Define time steps (e.g., past 10 readings)
time_steps = 10
X, y = create_sequences(data_values, time_steps)

# Print shape (should be 3D: samples, time_steps, 1)
print("X shape:", X.shape)  # (num_samples, time_steps, 1)
print("y shape:", y.shape)  # (num_samples,)
```

```python
# Define split index (80% training, 20% testing)
train_size = int(len(X) * 0.8)

# Split into training and testing sets
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Verify final shapes
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

```python
# Build GRU Model
model = Sequential([
    GRU(50, return_sequences=True, input_shape=(time_steps, 1)),  # GRU layer with 50 neurons
    GRU(50, return_sequences=False),  # Second GRU layer
    Dense(25),  # Fully connected layer
    Dense(1)    # Output layer (predict next water consumption value)
])

# Compile Model
model.compile(optimizer="adam", loss="mean_squared_error")

# Train Model
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```
