```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
water_usage=pd.read_csv('Water Usage/SOC Outlet-1/Table.csv',usecols=["Date","Time","Device Name","water_flow_in_cubic_meter"],parse_dates=["Date"],index_col="Date")
data=water_usage[::-1]
data.reset_index(inplace=True)
# Ensure data is in the correct order
data = data.sort_index()
data['Date'] = pd.to_datetime(data['Date'])  # Ensure datetime format
data.set_index('Date', inplace=True)  # Set timestamp as index


data.head(20)
```

```python

# Plot data
plt.figure(figsize=(10,5))
plt.plot(data["water_flow_in_cubic_meter"], label="Water Demand")
plt.xlabel("Time")
plt.ylabel("Water Consumption")
plt.title("Water Demand Over Time")
plt.legend()
plt.show()
```

```python
# Function to check stationarity
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print(f"ADF Statistic: {result[0]}")
    print(f"P-Value: {result[1]}")  
    if result[1] <= 0.05:
        print("Data is Stationary")
    else:
        print("Data is NOT Stationary")

# Check stationarity
check_stationarity(data["water_flow_in_cubic_meter"])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF (q) and PACF (p)
fig, ax = plt.subplots(1,2, figsize=(12,5))
plot_acf(data["water_flow_in_cubic_meter"].dropna(), lags=20, ax=ax[0])  # Finds q
plot_pacf(data["water_flow_in_cubic_meter"].dropna(), lags=20, ax=ax[1]) # Finds p
plt.show()
```

```python
import itertools
import warnings
warnings.filterwarnings("ignore")

# Define the range for p and q
p_values = range(0, 4)  # Try p from 0 to 3
d_values = [0]  # No differencing
q_values = range(0, 4)  # Try q from 0 to 3

# Grid search to find the best combination
best_aic = float("inf")
best_order = None

for p, d, q in itertools.product(p_values, d_values, q_values):
    try:
        model = ARIMA(data["water_flow_in_cubic_meter"], order=(p, d, q))
        model_fit = model.fit()
        aic = model_fit.aic  # Akaike Information Criterion (lower = better)
        if aic < best_aic:
            best_aic = aic
            best_order = (p, d, q)
    except:
        continue

print(f"Best ARIMA order: {best_order}")
print(f"p:{p},d:{d},q:{q}")
```

```python
model = ARIMA(data["water_flow_in_cubic_meter"], order=(p, d, q))

# Fit the model
model_fit = model.fit()

# Print model summary
print(model_fit.summary())
```

```python
# Forecast next 50 time steps
forecast_steps = 1000
forecast = model_fit.forecast(steps=forecast_steps) 

forecast_index = pd.date_range(start=data.index[-1], periods=forecast_steps, freq="6H")
#forecast_result = model_fit.get_forecast(steps=50, exog=future_exog)

# Convert to DataFrame
forecast_df = pd.DataFrame({"Timestamp": forecast_index, "Predicted_Water_Consumption": forecast})

# Plot forecast
plt.figure(figsize=(10,5))
plt.plot(data.index, data["water_flow_in_cubic_meter"], label="Actual Water Demand")
plt.plot(forecast_df["Timestamp"], forecast_df["Predicted_Water_Consumption"], label="Forecast", linestyle="dashed", color="red")
plt.xlabel("Time")
plt.ylabel("Water Consumption")
plt.title("ARIMA Water Demand Forecast")
plt.legend()
plt.show()
```

```python
# Train-Test Split (80% Train, 20% Test)
train_size = int(len(data) * 0.8)
train, test = data["water_flow_in_cubic_meter"].iloc[:train_size], data["water_flow_in_cubic_meter"].iloc[train_size:]

# Fit ARIMA on Training Data
model_train = ARIMA(train, order=(p, d, q))
model_train_fit = model_train.fit()

# Predict on Test Data
test_predictions = model_train_fit.forecast(steps=len(test))

# Compute RMSE
rmse = np.sqrt(mean_squared_error(test, test_predictions))
print(f"Test RMSE: {rmse:.2f}")

# Plot Actual vs. Predicted
plt.figure(figsize=(10,5))
plt.plot(test.index, test, label="Actual Water Demand")
plt.plot(test.index, test_predictions, label="Predicted Water Demand", linestyle="dashed", color="red")
plt.xlabel("Time")
plt.ylabel("Water Consumption")
plt.title("ARIMA Forecast vs Actual")
plt.legend()
plt.show()
```
