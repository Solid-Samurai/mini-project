```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

# Load dataset
df = pd.read_csv("Water Usage/SOC Outlet-1/Table.csv", usecols=["Date", "Time", "Device Name", "water_flow_in_cubic_meter"])
df = df[::-1]
df["Timestamp"] = pd.to_datetime(df["Date"] + " " + df["Time"], format="%d-%m-%Y %H:%M:%S")
df["Rounded_Timestamp"] = df["Timestamp"].dt.round("6H")
df.set_index("Rounded_Timestamp", inplace=True)
df.drop(columns=["Date", "Time", "Timestamp", "Device Name"], inplace=True)
df = df.resample("6h").mean()
df.rename(columns={'water_flow_in_cubic_meter': 'Water_Consumption'}, inplace=True)
df.interpolate(method="linear", inplace=True)

# Scaling
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
original_timestamps = df.index  # Save timestamps before preprocessing

# Sequence creation
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 10
X, y = create_sequences(scaled, seq_length)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# TimeSeriesSplit
tscv = TimeSeriesSplit(n_splits=5)
fold = 1
# Keep timestamps aligned with data
timestamps = df.index[seq_length:]

for train_idx, test_idx in tscv.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    test_timestamps = timestamps[test_idx]  # <-- Use timestamps

    # Build model
    model = Sequential([
        GRU(64, return_sequences=False, input_shape=(seq_length, 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=20, batch_size=16, verbose=0)

    # Predict and inverse scale
    y_pred = model.predict(X_test)
    y_test_inv = scaler.inverse_transform(np.column_stack([y_test]*df.shape[1]))[:, 0]
    y_pred_inv = scaler.inverse_transform(np.column_stack([y_pred]*df.shape[1]))[:, 0]

    # Plot with timestamps
    plt.figure(figsize=(12, 5))
    plt.plot(test_timestamps, y_test_inv, label="Actual")
    plt.plot(test_timestamps, y_pred_inv, label="Predicted", linestyle="--")
    plt.title(f"Fold {fold} - Prediction vs Actual")
    plt.xlabel("Timestamp")
    plt.ylabel("Water Flow")
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid(True)
    plt.show()

    fold += 1
```
