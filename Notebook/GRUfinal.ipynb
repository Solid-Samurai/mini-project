```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
# Load dataset
water_usage = pd.read_csv('Water Usage/SOC Outlet-1/Table.csv',
                          usecols=["Date", "Time", "Device Name", "water_flow_in_cubic_meter"])

water_lvl = pd.read_csv('Water Level/Kaveri+Water+Level.csv',
                          usecols=["Date", "Time", "Device Name","litres"])

# Reverse order if needed
data_flow = water_usage[::-1]
water_lvl["level_in_cubic_meter"]=water_lvl["litres"]*0.001
data_lvl=water_lvl[::-1]

  # Keep first occurrence

# Set index and ensure DateTime format
data_flow["Timestamp"] = pd.to_datetime(data_flow["Date"] + " " + data_flow["Time"],format="%d-%m-%Y %H:%M:%S")
data_lvl["Timestamp"] = pd.to_datetime(data_lvl["Date"] + " " + data_lvl["Time"],format="%d/%m/%Y %H:%M:%S")
data_lvl["Rounded_Timestamp"] = data_lvl["Timestamp"].dt.round("20min")

# Set the new Timestamp as index
data_flow.set_index("Timestamp", inplace=True)
data_lvl.set_index("Rounded_Timestamp", inplace=True)

# Drop the original "Date" and "Time" columns if no longer needed
data_flow.drop(columns=["Date", "Time","Device Name"], inplace=True)
data_lvl.drop(columns=["Date", "Time","Device Name","litres"], inplace=True)
#data_flow = data_flow.drop_duplicates(subset=["Rounded_Timestamp"], keep="first")

# Ensure daily frequency
# Resample water flow data to match water level frequency (20 minutes)
data_flow = data_flow.resample('20min').mean()
data_flow=data_flow.interpolate()

# Rename column
data_flow.rename(columns={'water_flow_in_cubic_meter': 'water_flow'}, inplace=True)
data_lvl.rename(columns={'level_in_cubic_meter': 'water_level'}, inplace=True)
#print(data_flow.isna().sum())
# Display first 20 rows
#print(data_flow.head(20))
data_flow.plot()
df=data_lvl.join(data_flow,how="inner")
df.drop(columns=["Timestamp"],inplace=True)
df.head(20)
df.plot()
```

```python

# Define motor control logic
low_level_threshold = 0.3  # Adjust based on normalized scale (0 to 1)
high_flow_threshold = 0.7

# Create a motor state column: 1 (ON) when level is low & flow increases
df['motor_state'] = ((df['water_level'] < low_level_threshold) & 
                            (df['water_flow'].diff() > high_flow_threshold)).astype(int)

# Normalize data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

# Convert to supervised learning format
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :2])  # Use water level & flow as input
        y.append(data[i+seq_length, 2])  # Predict motor state
    return np.array(X), np.array(y)

seq_length = 36  # Last 12 hours (20 min intervals)
X, y = create_sequences(scaled_data, seq_length)

# Split into train/test
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Build GRU Model
model = Sequential([
    GRU(64, return_sequences=True, input_shape=(seq_length, 2)),
    GRU(32),
    Dense(1, activation='sigmoid')  # Sigmoid for binary classification
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train Model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=15, batch_size=32)

# Predictions
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5).astype(int)  # Convert probability to 0 or 1

# Evaluate Model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Plot actual vs predicted motor state
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='Actual Motor State')
plt.plot(y_pred, label='Predicted Motor State', linestyle='dashed')
plt.legend()
plt.show()
```
